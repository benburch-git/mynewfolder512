## Question 5
This question runs through a linear regression example. We want to predict median house value based on the other variables.
#### a) First, load the `housing.csv` data set. Look at the data in some useful way. Why is linear regression appropriate here?

```r
#| message: FALSE

library(tidyverse)

housing <- read.csv("/content/housing.csv")

str(housing)
summary(housing)
head(housing)
```

We want to use linear regression because the response value is continuous.

#### b) Scale data and split it 75/25 training/testing. Set seed = 123.

```r
set.seed(123)

housing_scaled <- as.data.frame(scale(housing))

sample_index <- sample(1:nrow(housing_scaled), 0.75 * nrow(housing_scaled))
train <- housing_scaled[sample_index, ]
test <- housing_scaled[-sample_index, ]
```

#### c) Fit the model.

```r
model <- lm(median_house_value ~ ., data = train)

summary(model)
```

#### d) Make predictions on test data and show them in an actual vs. predicted plot.

```r
pred <- predict(model, newdata = test)

ggplot(data = NULL, aes(x = test$median_house_value, y = pred)) +
  geom_point(alpha = 0.6, color = "blue") +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed") +
  labs(
    title = "Actual vs Predicted Median House Value",
    x = "Actual Values",
    y = "Predicted Values"
  ) +
  theme_minimal()
```

#### e) Make a residuals plot.

```r
residuals <- test$median_house_value - pred

ggplot(data = NULL, aes(x = pred, y = residuals)) +
  geom_point(alpha = 0.6, color = "darkgreen") +
  geom_hline(yintercept = 0, color = "red", linetype = "dashed") +
  labs(
    title = "Residuals vs Predicted Values",
    x = "Predicted Values",
    y = "Residuals"
  ) +
  theme_minimal()
```

## Question 6
This question runs through a logistic regression example. We want to predict diabetes diagnosis based on the other variables.
#### a) First, load the `diabetes.csv` data set. Look at the data in some useful way. Why is logistic regression appropriate here?

```r
library(tidyverse)

diabetes <- read.csv("/content/diabetes.csv")

str(diabetes)
summary(diabetes)
head(diabetes)
```

#### b) Scale data and split it 75/25 training/testing. Set seed = 123.

```r
set.seed(123)

diabetes_scaled <- diabetes %>%
  mutate(Outcome = factor(Outcome, levels = c(0, 1))) %>%
  mutate(across(where(is.numeric), scale))


train_index <- sample(1:nrow(diabetes_scaled), 0.75 * nrow(diabetes_scaled))
train <- diabetes_scaled[train_index, ]
test <- diabetes_scaled[-train_index, ]
```

#### c) Fit the model.

```r
log_model <- glm(Outcome ~ ., data = train, family = binomial)

summary(log_model)
```

#### d) Make predictions on test data. Print a table with the number of true positives, false positives, true negatives, false negatives, and accuracy.

```r
prob <- predict(log_model, newdata = test, type = "response")
pred <- ifelse(prob > 0.5, 1, 0)

cm <- table(Predicted = pred, Actual = test$Outcome)
cm

TP <- cm["1","1"]
TN <- cm["0","0"]
FP <- cm["1","0"]
FN <- cm["0","1"]

accuracy <- (TP + TN) / sum(cm)

cat("True Positives:", TP, "\nFalse Positives:", FP,
    "\nTrue Negatives:", TN, "\nFalse Negatives:", FN,
    "\nAccuracy:", round(accuracy, 3))
```

#### e) Fit a LASSO-regularized logistic regression model. Again, set seed = 123. Which variables are the most important (which ones don't go to zero)? How does the LASSO model affect the accuracy?

```r
library(glmnet)

set.seed(123)

x <- model.matrix(Outcome ~ ., data = train)[, -1]
y <- train$Outcome

lasso_model <- cv.glmnet(x, y, family = "binomial", alpha = 1)

best_lambda <- lasso_model$lambda.min
best_lambda

coef_lasso <- coef(lasso_model, s = "lambda.min")
coef_lasso

important_vars <- rownames(coef_lasso)[which(coef_lasso != 0)]
important_vars

x_test <- model.matrix(Outcome ~ ., data = test)[, -1]
lasso_prob <- predict(lasso_model, newx = x_test, s = "lambda.min", type = "response")
lasso_pred <- ifelse(lasso_prob > 0.5, 1, 0)

cm_lasso <- table(Predicted = lasso_pred, Actual = test$Outcome)
cm_lasso

accuracy_lasso <- sum(diag(cm_lasso)) / sum(cm_lasso)
cat("LASSO Accuracy:", round(accuracy_lasso, 3))
```
